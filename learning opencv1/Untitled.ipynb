```python
import numpy as np, sys 
```

```python
A = cv2.imread('img.jpg')
B = cv2.imread('img.jpg')
```

```python
#生成高斯图像金字塔
G = A.copy()
gpA = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    print(G)
    gpA.append(G)
print(gpA)

G = B.copy()
gpB = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    gpB.append(G)
print(gpB)
```

```python
#对A产生laplacian
lpA = [gpA[5]]
print(type(lpA))
for i in range(5,0,-1):
    GE = cv2.pyrUp(gpA[i])
    #print(type(GE))
    #print('------------------------')
    #print(type(gpA))
    L = cv2.subtract(np.asarray(gpA[i-1]),GE)
    lpA.append(L)
```
